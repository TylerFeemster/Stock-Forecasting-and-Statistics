```python
import pandas as pd

df = pd.read_csv('vwap.csv')
df_vol = pd.read_csv('volume.csv')
df.index = list(df['Unnamed: 0'])
df_vol.index = list(df_vol['Unnamed: 0'])
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df_vol.drop(['Unnamed: 0'], axis=1, inplace=True)
display(df.describe())
df_vol.describe()
```

```python
import numpy as np

df_log = df.copy()
df_log.iloc[:,:] = df.iloc[:,:].apply(lambda x: np.log1p(x))
df_log.describe()
```

```python
df_diff = df_log - df_log.shift()
df_diff.describe()
```

```python
corr = df_diff.corr()
display(corr.mean(axis=1).sort_values())
```

```python
for idx, row in corr.iterrows():
    for col in corr.columns:
        if idx != col and row[col] > 0.95:
            print(f"{col:6}{idx:6}{row[col]:.5f}")
```

```python
df_diff.drop(['GOOGL', 'FOXA', 'NWSA'], axis=1, inplace=True)
corr = df_diff.corr()
```

## Intra-Sector

```python
df_sp = pd.read_csv('s&p500.csv')
sectors = list(df_sp['GICS Sector'].value_counts().index)
```

```python
print(df_sp.index)

by_industry = {}
for sector in sectors:
    stocks = list(df_sp.loc[df_sp['GICS Sector'] == sector]['Symbol'])
    stks = [stk for stk in stocks if stk in corr.columns]
    by_industry[sector] = stks
    print(f"{sector:25}", stks)
```

```python
series = pd.Series()
for sector in sectors:
    stks = by_industry[sector]
    series[sector] = (corr.loc[stks, stks].sum(0).sum(0) - len(stks)) / (len(stks) * (len(stks) - 1))

series.sort_values(inplace=True)
```

## Inter-Sector

```python
df_by_sector = pd.DataFrame()
for sector in sectors:
    stks = by_industry[sector]
    df_by_sector[sector] = (df[stks] * df_vol[stks]).sum(axis=1) / df_vol[stks].sum(axis=1)

df_sect_log = df_by_sector.iloc[:,:].apply(lambda x: np.log1p(x))
df_sector_diffs = df_sect_log - df_sect_log.shift()

display(df_sector_diffs.describe())

sector_corr = df_sector_diffs.corr()

sector_corr['sort_col'] = sector_corr.sum(axis=0)

sector_corr.sort_values(by='sort_col', inplace=True, ascending=False)

sector_corr.drop('sort_col', axis=1, inplace=True)

# permute columns to match sorted ordering
sector_corr = sector_corr[sector_corr.index]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(series, orient='h')
plt.title('Correlation of Stocks within Industries')
plt.show()

sns.heatmap(sector_corr, vmax=0.2, vmin=-0.1, center=0, cmap='coolwarm')
plt.title('Correlation of Stocks between Industries')
plt.show()
```

### Statistical Significance

```python
from scipy.stats import t
import matplotlib.pyplot as plt

def p_value(n, dataframe: pd.DataFrame) -> pd.DataFrame:
    p_val = pd.DataFrame(index=dataframe.index, columns=dataframe.columns)
    for sector in sectors:
        t_stat = dataframe[sector] * np.sqrt((n - 2) / (1 - dataframe[sector]**2))
        p_val[sector] = 2 * t.sf(np.abs(t_stat), df=n-2)
        p_val.index = dataframe.index
    return p_val

pval = p_value(len(df_sector_diffs), sector_corr)
pval['sort_col'] = pval.sum(axis=0)
pval.sort_values(by='sort_col', inplace=True, ascending=False)
pval.drop('sort_col', axis=1, inplace=True)
pval = pval[pval.index]

sns.heatmap(pval, vmax=0.02, vmin=0, center=0.01, cmap=plt.cm.coolwarm.reversed())
plt.title('P-Value of Correlation between Industries')
plt.show()
```

```python
result = pd.concat([
    pd.Series(
        [df_sector_diffs[idx].corr(df_sector_diffs[col].shift()) for idx in df_sector_diffs.columns], 
        index=df_sector_diffs.columns) for col in df_sector_diffs.columns], axis=1)
result.columns = df_sector_diffs.columns

sns.heatmap(result, vmax=0.1, vmin=-0.1, center=0, cmap='coolwarm')
plt.title('Correlation of Stocks between Industries Shifted by 1 Day')
plt.show()

pval2 = p_value(len(df_sector_diffs) - 1, result)

pval2['sort_col'] = pval2.sum(axis=0)
pval2.sort_values(by='sort_col', inplace=True, ascending=False)
pval2.drop('sort_col', axis=1, inplace=True)
pval2 = pval2[pval2.index]

sns.heatmap(pval2, vmax=0.1, vmin=0, center=0.05,
            cmap=plt.cm.coolwarm.reversed())
plt.title('P-Values of Correlations of Stocks between Industries Shifted by 1 Day')
plt.show()
```
